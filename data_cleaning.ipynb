```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to generate uniformly distributed data on a 3D sphere
def generate_sphere_data(num_points):
    u = np.random.uniform(0, 1, num_points)
    v = np.random.uniform(0, 1, num_points)
    
    theta = 2 * np.pi * u
    phi = np.arccos(2 * v - 1)
    
    x1 = np.sin(phi) * np.cos(theta)
    x2 = np.sin(phi) * np.sin(theta)
    x3 = np.cos(phi)
    
    sphere_data = pd.DataFrame({'x1': x1, 'x2': x2, 'x3': x3})
    return sphere_data

# Function to generate uniformly distributed data on a 3D parabolic surface
def generate_parabolic_data(num_points, max_radius=1):
    u = np.random.uniform(0, 1, num_points)
    theta = np.random.uniform(0, 2 * np.pi, num_points)
    
    r = max_radius * np.sqrt(u)
    x1 = r * np.cos(theta)
    x2 = r * np.sin(theta)
    x3 = x1**2 + x2**2
    
    parabolic_data = pd.DataFrame({'x1': x1, 'x2': x2, 'x3': x3})
    return parabolic_data

# Number of data points
num_points = 50000

# Generate sphere data
sphere_data = generate_sphere_data(num_points)
# Save sphere data to CSV
sphere_data.to_csv('sphere.csv', index=False)

# Generate parabolic data
parabolic_data = generate_parabolic_data(num_points)
# Save parabolic data to CSV
parabolic_data.to_csv('parabolic.csv', index=False)

# Plot histograms for all features in both datasets
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Feature names
features = ['x1', 'x2', 'x3']
colors = ['blue', 'red']

for i, feature in enumerate(features):
    axes[i].hist(sphere_data[feature], bins=30, alpha=0.5, color=colors[0], label='Sphere Data')
   # axes[i].hist(parabolic_data[feature], bins=30, alpha=0.5, color=colors[1], label='Parabolic Data')
    axes[i].set_title(f'Histogram of {feature}')
    axes[i].legend()

# 3D scatter plots for data distribution
fig = plt.figure(figsize=(14, 7))

# 3D scatter plot for sphere data
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(sphere_data['x1'], sphere_data['x2'], sphere_data['x3'], color='blue', alpha=0.6)
ax1.set_title('3D Scatter Plot of Sphere Data')
ax1.set_xlabel('x1')
ax1.set_ylabel('x2')
ax1.set_zlabel('x3')
'''
# 3D scatter plot for parabolic data
ax2 = fig.add_subplot(122, projection='3d')
ax2.scatter(parabolic_data['x1'], parabolic_data['x2'], parabolic_data['x3'], color='red', alpha=0.6)
ax2.set_title('3D Scatter Plot of Parabolic Data')
ax2.set_xlabel('x1')
ax2.set_ylabel('x2')
ax2.set_zlabel('x3')

plt.tight_layout()
plt.savefig('hi.pdf')
plt.show()'''
```
