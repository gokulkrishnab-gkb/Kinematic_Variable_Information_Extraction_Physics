```python
import pandas as pd

# Load the CSV file
file_path = 'data/particle_extract.csv'
data = pd.read_csv(file_path)

# Define the PID values to filter
pid_values = [11,13]

# Group the data by Event_Number and filter based on PID values
filtered_data = data[data['PID'].isin(pid_values)]

# Save the filtered data to a new CSV file
output_file_path = 'data/filtered_particle_extract.csv'
filtered_data.to_csv(output_file_path, index=False)

print(f"Filtered data saved to {output_file_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the filtered CSV file
filtered_file_path = 'data/filtered_particle_extract.csv'
filtered_data = pd.read_csv(filtered_file_path)

# Filter the data to include only PT values greater than 25 and less than or equal to 1000
filtered_data =  filtered_data[(filtered_data['PT'] > 25) & (filtered_data['PT'] <= 1000)]

# Plot a histogram of the PT values
plt.figure(figsize=(10, 6))
plt.hist(filtered_data['PT'], bins=100, edgecolor='b', alpha=1, histtype='step')
plt.yscale('log')
plt.xlabel('PT')
plt.ylabel('Frequency')
plt.title('Histogram of PT values from filtered_particle_extract.csv (25 < PT <= 1000)')
plt.yticks([ 1, 10, 100])

# Show the plot
plt.show()
```
