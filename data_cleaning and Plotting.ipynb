```python
import pandas as pd

# Load the CSV file
file_path = 'data/particle_extract.csv'
data = pd.read_csv(file_path)

# Define the PID values to filter
pid_values = [11,13]

# Group the data by Event_Number and filter based on PID values
filtered_data = data[data['PID'].isin(pid_values)]

# Save the filtered data to a new CSV file
output_file_path = 'data/filtered_particle_extract.csv'
filtered_data.to_csv(output_file_path, index=False)

print(f"Filtered data saved to {output_file_path}")
```

```python
import pandas as pd

# Load the CSV file
file_path = 'data/particle_extract_elementary.csv'
data = pd.read_csv(file_path)

# Define the PID values to filter
pid_values = [11,13]

# Group the data by Event_Number and filter based on PID values
filtered_data = data[data['PID'].isin(pid_values)]

# Save the filtered data to a new CSV file
output_file_path = 'data/filtered_particle_extract_elementary.csv'
filtered_data.to_csv(output_file_path, index=False)

print(f"Filtered data saved to {output_file_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the first filtered CSV file
filtered_file_path_1 = 'data/filtered_particle_extract.csv'
filtered_data_1 = pd.read_csv(filtered_file_path_1)

# Filter the data to include only PT values greater than 25 and less than or equal to 1000
filtered_data_1 = filtered_data_1[(filtered_data_1['PT'] > 25) & (filtered_data_1['PT'] <= 1000)]

# Load the second filtered CSV file
filtered_file_path_2 = 'data/filtered_particle_extract_elementary.csv'
filtered_data_2 = pd.read_csv(filtered_file_path_2)

# Filter the data to include only PT values greater than 25 and less than or equal to 1000
filtered_data_2 = filtered_data_2[(filtered_data_2['PT'] > 25) & (filtered_data_2['PT'] <= 1000)]

# Plot histograms of the PT values on the same canvas
plt.figure(figsize=(10, 6))
plt.hist(filtered_data_1['PT'], bins=100, edgecolor='b', alpha=1.0, histtype='step', label='Form Factor: Gamma/M= 1%')
plt.hist(filtered_data_2['PT'], bins=100, edgecolor='r', alpha=1.0, histtype='step', label='Elementary Higgs')
plt.yscale('log')
plt.xlabel('PT')
plt.ylabel('Frequency')
plt.title('Histogram of PT values (25 < PT <= 1000)')
plt.yticks([1, 10, 100])
plt.legend()

# Show the plot
plt.show()
```
